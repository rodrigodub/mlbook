### Load data using Pandas

```python
import os
import pandas as pd

pd.options.mode.chained_assignment = None  # default='warn'
# source https://stackoverflow.com/questions/20625582/how-to-deal-with-settingwithcopywarning-in-pandas 

SAMPLESPATH = os.path.join('datasets', 'spif')
```

```python
# function to load data

def loaddata(path, file):
    filetoretrieve = os.path.join(path, file)
    return pd.read_csv(filetoretrieve)
```

##### 2.1 first with a subsample

```python
# test, take a subset of samples
subsamples = samples[0:1000]
subsamples
# subsamples['HOLEID'].unique()
```

```python
# step 1 : merge everything from litho
mergesubsamplelithofull = pd.merge(subsamples, litho,  how='left', left_on=['HOLEID'], right_on = ['HOLEID'])
mergesubsamplelithofull
```

```python
# step 2 : filter for where Sample interval is related to Litho interval
subsamplelitho = mergesubsamplelithofull[(mergesubsamplelithofull['FROM_x'] <= mergesubsamplelithofull['TO_y'])
                                   & (mergesubsamplelithofull['TO_x'] >= mergesubsamplelithofull['FROM_y'])]
subsamplelitho
```

```python
subsamplelitho['LITH'].unique()
```
